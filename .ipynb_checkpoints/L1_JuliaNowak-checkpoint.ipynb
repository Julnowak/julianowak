Zadanie rozpoczęłam od zaimportowania odpowiednich bibliotek, niezbędnych do wykonania zadania.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

Następnie zdefiniowałam funkcję wskazaną w treści zadania.

```python
def f(x):
    return x**2 + 5
```

W następnym kroku określiłam zakresy danych - przyjęłam w każdym przypadku, że krok jest równy 0.1.\
Warunek kończący jest większy o 0.1 po to, aby na wykresie zmieścił się pełny zakres danych \
(funkcja np.arange nie uwzględnia czasami wartości końcowej).

```python
x1 = np.arange(-1,1.1,0.1)
x2 = np.arange(-6,6.1,0.1)
x3 = np.arange(0,5.1,0.1)
```

Na końcu, przystąpiłam do stworzenia wykresu funkcji z oznaczonymi przedziałami.

```python
# Wszystkie zakresy na jednym wykresie
# Ustalenie rozmiaru figury
plt.rcParams["figure.figsize"] = (10,10)

# Utworzenie wykresów dla poszczególnych zestawów danych
# Wykresom nadano kolory i pogrubiono linie dla lepszej czytelności
plt.plot(x1,f(x1),label='Arguments from -1 to 1',linewidth=10,color='orange')
plt.plot(x2,f(x2),label='Arguments from -6 to 6',color='red',linewidth=2)
plt.plot(x3,f(x3),label='Arguments from 0 to 5',color='blue',linewidth=5)

# Dodano tytuł
plt.title('Function f(x) = x^2 + 5')

# Dodano podpisy osi
plt.xlabel('X axis')
plt.ylabel('Y axis')

#Dodano legendę
plt.legend()

# Dodano siatkę
plt.grid()

# Pokazano wykres
plt.show()
```

Wykresy dla każdego z zakresów osobno:

```python
# Wykres 1
# Wykresowi nadano kolor dla lepszej czytelności
plt.plot(x1,f(x1),label='f(x) = x^2 + 5',color='orange')

# Dodano tytuł
plt.title('Arguments from -1 to 1')

# Dodano podpisy osi
plt.xlabel('X axis')
plt.ylabel('Y axis')

#Dodano legendę
plt.legend()

# Dodano siatkę
plt.grid()

# Pokazano wykres
plt.show()
```

```python
# Wykres 2
# Wykresowi nadano kolor dla lepszej czytelności
plt.plot(x2,f(x2),label='f(x) = x^2 + 5',color='red')

# Dodano tytuł
plt.title('Arguments from -6 to 6')

# Dodano podpisy osi
plt.xlabel('X axis')
plt.ylabel('Y axis')

#Dodano legendę
plt.legend()

# Dodano siatkę
plt.grid()

# Pokazano wykres
plt.show()
```

```python
# Wykres 3
# Wykresowi nadano kolor dla lepszej czytelności

plt.plot(x3,f(x3),label='f(x) = x^2 + 5',color='blue')

# Dodano tytuł
plt.title('Arguments from 0 to 5')

# Dodano podpisy osi
plt.xlabel('X axis')
plt.ylabel('Y axis')

#Dodano legendę
plt.legend()

# Dodano siatkę
plt.grid()

# Pokazano wykres
plt.show()
```
