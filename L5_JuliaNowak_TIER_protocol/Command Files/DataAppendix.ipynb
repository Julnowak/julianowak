# Przedstawienie danych pogodowych na wykresie dla wybranych lat

## Wczytanie ostatecznej Dataframe

```python
from cleaning_data import *
import matplotlib.pyplot as plt

print(df_final)
```

## Kod tworzący wykres, przedstawiający zmiany temperatur w latach 1955 - 1979.

```python
fig, ax = plt.subplots(figsize=((30,20)))


df_final.plot(kind = 'line',
        x = 'date',
        y = 'tmax',
        color = 'red',ax = ax)

  
df_final.plot(kind = 'line',x = 'date',
        y = 'tmin',
        color = 'black',ax = ax)
  

plt.title('Zestawienie temperatury maksymalnej i minimalnej w Meksyku w latach 1955-1979',fontsize=30)
plt.legend(fontsize=20)
ax.set_xlabel('Data', fontsize=20)
ax.set_ylabel('Temperatura [K]', fontsize=20)
plt.rcParams.update({'font.size': 10})  

plt.show()
```

```python
fig, ax = plt.subplots(figsize=((30,20)))

df_final[0:278].plot(kind = 'line',
        x = 'date',
        y = 'tmax',
        color = 'red',ax = ax)

  
df_final[0:278].plot(kind = 'line',x = 'date',
        y = 'tmin',
        color = 'black',ax = ax)
  
plt.title('Zestawienie temperatury maksymalnej i minimalnej w Meksyku w 1955 roku',fontsize=30)
plt.legend(fontsize=20)
ax.set_xlabel('Data', fontsize=20)
ax.set_ylabel(u'Temperatura [K]', fontsize=20)
plt.rcParams.update({'font.size': 10})  

plt.show()
```
