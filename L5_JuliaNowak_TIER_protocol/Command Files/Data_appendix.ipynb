### Dane dotyczące temperatur

```python
from cleaning_data import *
from cleaning_opady import *
import matplotlib.pyplot as plt

print(df_final)
```

```python
# Informacje o danych o temperaturze maksymalnej
mean = df_final['tmax'].mean()
print('średnia wartość: ' + str(mean))

stdd = df_final['tmax'].std()
print('Odchylenie standardowe: ' + str(stdd))

mini = df_final['tmax'].min()
print('Wartość minimalna: ' + str(mini))

maxi = df_final['tmax'].max()
print('Wartość maksymalna: ' + str(maxi))

kw25 = df_final['tmax'].quantile(0.25)
print('Kwantyl 25%: ' + str(kw25))

mediana = df_final['tmax'].quantile()
print('Mediana: ' + str(mediana))

kw75 = df_final['tmax'].quantile(0.75)
print('Kwantyl 75%: ' + str(kw75))

hist = df_final['tmax'].hist()
```

```python
# Informacje o danych o temperaturze minimalnej
mean = df_final['tmin'].mean()
print('średnia wartość: ' + str(mean))

stdd = df_final['tmin'].std()
print('Odchylenie standardowe: ' + str(stdd))

mini = df_final['tmin'].min()
print('Wartość minimalna: ' + str(mini))

maxi = df_final['tmin'].max()
print('Wartość maksymalna: ' + str(maxi))

kw25 = df_final['tmin'].quantile(0.25)
print('Kwantyl 25%: ' + str(kw25))

mediana = df_final['tmin'].quantile()
print('Mediana: ' + str(mediana))

kw75 = df_final['tmin'].quantile(0.75)
print('Kwantyl 75%: ' + str(kw75))

hist = df_final['tmin'].hist()
```

## Kod tworzący wykres, przedstawiający zmiany temperatur w latach 1955 - 1980.

```python
fig, ax = plt.subplots(figsize=((30,20)))


df_final.plot(kind = 'line',
        x = 'date',
        y = 'tmax',
        color = 'red',ax = ax)

  
df_final.plot(kind = 'line',x = 'date',
        y = 'tmin',
        color = 'black',ax = ax)
  

plt.title('Zestawienie temperatury maksymalnej i minimalnej w Meksyku w latach 1955-1979',fontsize=30)
plt.legend(fontsize=20)
ax.set_xlabel('Data', fontsize=20)
ax.set_ylabel('Temperatura [K]', fontsize=20)
plt.rcParams.update({'font.size': 10})  
plt.grid()

plt.show()
```
